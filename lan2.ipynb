```python
def check_l_in_df(l, df):
    for n in l:
        if n not in df:
            return False
    return True
```

CÂU 4:

```python
from sklearn.metrics import mean_absolute_error, r2_score
from sklearn.linear_model import LinearRegression
```

```python
MLR = LinearRegression()
MLR.fit(df_train[['c1','d1','f1','c2','d2','f2','f3','f4','f5','f6']].values,df_train['limit'].values)
y_MLR_pred = MLR.predict(df_test[['c1','d1','f1','c2','d2','f2','f3','f4','f5','f6']].values)
mae_MLR = mean_absolute_error(df_test['limit'].values, y_MLR_pred)
r2_MLR = r2_score(df_test['limit'].values, y_MLR_pred)
print("MAE of MLR: ", mae_MLR)
print("R2 of MLR: ", r2_MLR)
```

CÂU 7:

```python
def find_max(a):
    max_num = -1
    for num in a:
        if num > max_num:
            max_num = num
    return max_num
```

```python
KNN.predict_proba(df_train[['c1','d1','f1']].values)
```

```python
knn_features = pd.DataFrame(KNN.predict_proba(df_train[['c1','d1','f1']].values)).apply(find_max, axis=1)
gnb_feature = pd.DataFrame(GNB.predict_proba(df_train[['c2','d2','f2']].values)).apply(find_max,axis = 1)
```
