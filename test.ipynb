```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error
import numpy as np
import pandas as pd
df = data
# =============================
# 1️⃣ Đọc dữ liệu
# =============================
# Giả sử df chứa các cột: f1, f2, f3, f4, f5, f6, limit
# df = pd.read_csv("data.csv")

X = df[['f1','f2','f3','f4','f5','f6']].values
y = df['limit'].values

# Chia tập train/test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=10)

# =============================
# 2️⃣ Trích xuất đặc trưng
# =============================

# --- KNN feature ---
knn = KNeighborsClassifier(n_neighbors=20)
knn.fit(X_train[:, 0:2], y_train)
knn_train_feat = knn.predict_proba(X_train[:, 0:2]).max(axis=1).reshape(-1,1)
knn_test_feat  = knn.predict_proba(X_test[:, 0:2]).max(axis=1).reshape(-1,1)

# --- GaussianNB feature ---
gnb = GaussianNB()
gnb.fit(X_train[:, 1:3], y_train)
gnb_train_feat = gnb.predict_proba(X_train[:, 1:3]).max(axis=1).reshape(-1,1)
gnb_test_feat  = gnb.predict_proba(X_test[:, 1:3]).max(axis=1).reshape(-1,1)

# --- Giữ nguyên f3–f6 ---
f_train = X_train[:, 2:6]
f_test  = X_test[:, 2:6]

# Gộp thành bộ đặc trưng hybrid
X_train_hybrid = np.hstack([knn_train_feat, gnb_train_feat, f_train])
X_test_hybrid  = np.hstack([knn_test_feat,  gnb_test_feat,  f_test])

# =============================
# 3️⃣ Polynomial Feature Expansion
# =============================
poly = PolynomialFeatures(degree=2, include_bias=True)
X_train_poly = poly.fit_transform(X_train_hybrid)
X_test_poly  = poly.transform(X_test_hybrid)

# =============================
# 4️⃣ Hồi quy tuyến tính
# =============================
mlr = LinearRegression()
mlr.fit(X_train_poly, y_train)

y_pred_train = mlr.predict(X_train_poly)
y_pred_test  = mlr.predict(X_test_poly)

# =============================
# 5️⃣ Đánh giá mô hình
# =============================
r2_train = r2_score(y_train, y_pred_train)
mae_train = mean_absolute_error(y_train, y_pred_train)

r2_test = r2_score(y_test, y_pred_test)
mae_test = mean_absolute_error(y_test, y_pred_test)

print("==== Hybrid Model Evaluation ====")
print(f"R² (Train): {r2_train:.4f}")
print(f"MAE (Train): {mae_train:.4f}")
print(f"R² (Test): {r2_test:.4f}")
print(f"MAE (Test): {mae_test:.4f}")
```

```python
x_input = np.array([[-1.0, 9640.0, 15134.0, -2.0, 7404.0, 0.0, 7002.0, 8167.0, 3996.0, 2000.0]])

# Trích xuất các feature như pipeline HM
# KNN feature
knn_feature = knn.predict_proba(x_input[:, [0, 1]]).max(axis=1).reshape(-1,1)

# GNB feature
gnb_feature = gnb.predict_proba(x_input[:, [3, 4]]).max(axis=1).reshape(-1,1)

# Giữ nguyên f3–f6
f_features = x_input[:, [6,7,8,9]]

# Kết hợp lại
X_input_hybrid = np.hstack([knn_feature, gnb_feature, f_features])

# Biến đổi Polynomial
X_input_poly = poly.transform(X_input_hybrid)

# Dự báo limit
limit_pred = mlr.predict(X_input_poly)
print("Giá trị dự báo limit:", limit_pred[0])
```
