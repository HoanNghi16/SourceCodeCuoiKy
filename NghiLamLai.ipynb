```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
#Giai đoạn 1
df = pd.read_excel('Dethi_data.xlsx')
```

```python
#MLR giai đoạn 1
X_1 = ['c1', 'd1', 'f1', 'c2', 'd2', 'f2', 'f3', 'f4', 'f5', 'f6']
y_1 = ['limit']
```

```python
from sklearn.model_selection import train_test_split
df_train, df_test = train_test_split(df, test_size=0.1, random_state=10)
```

```python
df_train
```

```python
X_train_1 = df_train[X_1].copy()
y_train_1 = df_train[y_1].copy()
X_test_1 = df_test[X_1].copy()
y_test_1 = df_test[y_1].copy()
```

```python
from sklearn.linear_model import LinearRegression
MLR1 = LinearRegression()
```

```python
MLR1.fit(X_train_1.values,y_train_1.values)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
mae1 = mean_absolute_error(y_test_1, MLR1.predict(X_test_1))
r21 = r2_score(y_test_1, MLR1.predict(X_test_1))
print(mae1)
print(r21)
```

```python
KNN_X_train = df_train[['c1', 'd1', 'f1']].copy()
KNN_y_train = df_train['drop'].copy()
KNN_X_test = df_test[['c1','d1','f1']].copy()
KNN_y_test = df_test['drop'].copy()
```

```python
#Câu 6
from sklearn.metrics import accuracy_score


KNN1 = KNeighborsClassifier(n_neighbors=20)
KNN1.fit(KNN_X_train.values, KNN_y_train.values)
KNN1_y_pred = KNN1.predict(KNN_X_train.values)
print(accuracy_score(KNN_y_train, KNN1_y_pred))
```

```python
#Câu 7
knn_feature = KNN1.predict_proba(KNN_X_train.values).max(axis=1)
```

```python
round(np.mean(knn_feature),2)
```

```python
GNB = GaussianNB()

GNB_X_train = df_train[['c2','d2','f2']].copy()
GNB_y_train = df_train['delete'].copy()
GNB_X_test = df_test[['c2','d2','f2']].copy()
GNB_y_test = df_test['delete'].copy()

GNB.fit(GNB_X_train.values, GNB_y_train.values)
gnb_feature = GNB.predict_proba(GNB_X_train.values).max(axis=1)
```

```python
gnb_feature
```

```python
X_train_2 = pd.DataFrame({'knn_feature':knn_feature,
                         'gnb_feature':gnb_feature,
                           'f3': df_train.f3.values,
                             'f4': df_train.f4.values,
                             'f5': df_train.f5.values,
                             'f6': df_train.f6.values})

y_train_2 = df_train['limit']
X_test_2 = pd.DataFrame({'knn_feature':KNN1.predict_proba(KNN_X_test.values).max(axis=1),
                         'gnb_feature':GNB.predict_proba(GNB_X_test.values).max(axis=1),
                           'f3': df_test.f3.values,
                             'f4': df_test.f4.values,
                             'f5': df_test.f5.values,
                             'f6': df_test.f6.values})


y_test_2 = df_test['limit']
```

```python
#Câu 8
arr = X_train_2.values.astype(float)
print("Min:", np.min(arr))
print("Mean:", np.mean(arr))
print("Var:", np.var(arr))
print("Max:", np.max(arr))
print("-------------------------------------------------")
```

```python
from sklearn.preprocessing import PolynomialFeatures
Poly = PolynomialFeatures(degree=2, include_bias=True)
X_train_2 = Poly.fit_transform(X_train_2.values)
X_test_2 = Poly.transform(X_test_2.values)
```

```python
#Câu 9:
MLR2 = LinearRegression()

MLR2.fit(X_train_2,y_train_2)
MLR2_y_pred = MLR2.predict(X_test_2)

mae2 = mean_absolute_error(y_test_2, MLR2_y_pred)
r22 = r2_score(y_test_2, MLR2_y_pred)

print(f'mae1: {mae1}')
print(f'mae2: {mae2}')
print('---------------------')
print(f'r21: {r21}')
print(f'r22: {r22}')
# a. MAE của hybrid là 82906.8479776912 -> sai
# b. Mô hình Hybrid và MLR tương đương nhau ->  sai
# c. R-squared của hybrid là 0.5747749190894 -> sai
# d. Sai số MAE trên Hybrid nhỏ hơn MLR -> sai
# e. Mô hình MLR tốt hơn -> đúng
# f. R-Squared của Hybrid là 0.36049 -> sai
# g. Sai số đánh giá của MAE trên mô hình Hybrid lớn hơn MLR -> đúng
# h. MAE cảu hybrid là 6063.2127933 -> sai
```
